# Testing

In this part of the tutorial, we'll learn how to test our code.

First, run a test by clicking on your terminal and entering

```bash
$ make test
```

If all goes well, you should get a message telling you your test passed.

When we run `make test`, 100 fake users run through our app and respond to questions randomly. This is a tremedously effective way to find and fix bugs.

Now, click on your terminal and enter

```bash
$ make testserve
```

This opens a *code coverage report* in your simple browser. A code coverage report tells us which parts of our code were tested. In your simple browser, click on `src/my_survey.py`. If all goes well, we should see 100% coverage in our main code file(s). Right now, we should see that every line in `src/my_survey.py` is highlighted in green, meaning that every line was tested by at least one of our fake users. If a line of code was highlighted in red, it would mean it was tested, which isn't good.

You should also see a new folder named `data` in your file explorer that contains a file `test.csv`. These are the data generated by your test users. After running your tests, I recommend checking `test.csv` to see that the data file looks the way you want it to.

By default, the test users respond to questions randomly. But, sometimes, we don't want our users to respond randomly. For example, suppose you have an `Input` asking the user some question to which there is a correct response, and the user can't continue until entering it. (We'll cover validation later.) Our test user, responding randomly, would be stuck on this page forever. Below, we'll learn how to set test users' default responses to avoid situations like this.

Start by running the cell below to create a test app.

````python
import os

from hemlock import User, Page, create_test_app
from hemlock.questions import Check, Input, Label, Select

os.environ.pop("GITPOD_HOST", None)

app = create_test_app()
````

Run the cell below and notice how the test user responds randomly. Notice a line in the output logs that says `"test response: '<some-random-response>'"`.

````python
def seed():
    return [
        Page(
            Input("What's your name?")
        ),
        Page(
            Label("Goodbye!")
        )
    ]


_ = User.make_test_user(seed).test_request()
````

We can change the default response by passing `test_response="<some-test-response>"`.

In the example below, we ask the user for their name and change the test user's repsonse to `"Seneca"`.

````python
def seed():
    return [
        Page(
            Input(
                "What's your name?",
                test_response="Seneca"
            )
        ),
        Page(
            Label("Goodbye!")
        )
    ]


_ = User.make_test_user(seed).test_request()
````

The test user can also respond with multiple choices for a multiple choice question.

````python
def seed():
    return [
        Page(
            Check(
                "Which of these songs do you like?",
                ["Here comes the sun", "Hard day's night", "Yesterday"],
                multiple=True,
                test_response=["Here comes the sun", "Yesterday"]
            )
        ),
        Page(
            Label("Goodbye!")
        )
    ]


_ = User.make_test_user(seed).test_request()
````
